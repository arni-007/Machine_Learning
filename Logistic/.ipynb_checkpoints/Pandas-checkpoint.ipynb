```python
# 6. Identify the number of missing values in each column.
# 7. Calculate the percentage of missing values per column.
# 8. Identify the column with the highest missing values and justify whether it should be dropped.
# 9. Handle missing values in the Age column.
# 10. Handle missing values in the Embarked column.
```

```python
df.isnull().sum()
```

```python
missing_per=df.isnull().sum() * 100 / len(df)
missing_per
```

```python
max_miss_col=df.isnull().sum().idxmax()
max_miss_count=df.isnull().sum().max()
print("max missing column-",max_miss_col)
print("max missing count-",max_miss_count)

miss_percent_max = (max_miss_count / len(df)) * 100

if(miss_percent_max>40):
    print("The column should be dropped as it has more than 40% missing values")
    df.drop(columns=[max_miss_col])
df
```

```python
mean_age=df['Age'].mean()
print(mean_age)
df['Age'].fillna(mean_age,inplace=True)

df
```

```python
mod_emb=df['Embarked'].mode()
df['Embarked'].fillna(mod_emb,inplace=True)
```

```python
# 14. Convert categorical columns into numerical format.
# 15. Apply Label Encoding to the Sex column.
# 16. Apply One-Hot Encoding to the Embarked column.
# 17. Justify the encoding techniques used.
```

```python
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df = pd.get_dummies(df, columns=["Embarked"])
df
```

```python
# 18. Detect outliers in the Fare column using the IQR method.
# 19. Visualize outliers using a boxplot.
# 20. Handle the detected outliers and justify your approach.
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[(df['Fare'] < Q1 - 1.5 * IQR) | (df['Fare'] > Q3 + 1.5 * IQR)]
outliers
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['Fare'])
plt.title('Boxplot of Fare')
plt.ylabel('Fare')
plt.show()
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df['Fare'] = df['Fare'].clip(lower=lower_limit, upper=upper_limit)
df
```

```python
# 21. Extract passenger titles from the Name column.
# 22. Group rare titles into a single category.
# 23. Create a new feature called FamilySize.
# 24. Drop columns that are not useful for prediction and justify.
```

```python
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
df
```

```python
common_titles = ["Mr", "Mrs", "Miss", "Master"]
df["Title"] = df["Title"].apply(lambda x: x if x in common_titles else "Rare")
df
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df
```

```python
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df
```

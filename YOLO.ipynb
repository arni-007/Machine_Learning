```
print(os.listdir("/content/dataset/cricket_ball_data"))
```

```
data_yaml = """
path: /content/dataset/cricket_ball_data
train: train/images
val: valid/images

names:
  0: cricket_ball
"""

with open("/content/data.yaml", "w") as f:
    f.write(data_yaml)

print("data.yaml created")
```

```
model = YOLO("yolov8n.pt")

model.train(
    data="/content/data.yaml",
    epochs=20,
    imgsz=640
)
```

```
model = YOLO("/content/runs/detect/train/weights/best.pt")

model.predict(
    source="/content/dataset/cricket_ball_data/test/images",
    save=True,
    show=True
)
```
